#  Medidas básicas de la estadística descriptiva

##  Medidas de centralización

```python
import pandas as pd
import numpy as np
import random
import scipy.stats as sp
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# Generar números aleatorios

random.seed(0)
X = np.random.choice(a = 1000, size = 10000, replace = True)
X
```

```python
# Cálculo media

mean = np.mean(X)
mean
```

```python
# Cálculo mediana

np.median(X)
```

```python
# Cálculo moda

sp.mode(X)
```

```python
len(np.where(X == 166)[0])
```

```python
# Percencil 100 = p1 0 = p0, 0.25 1er cuantil 0.75 3er cuantil

np.percentile(X, 25) # return 0.25 percentile, e.g 1er cuantil.
```

##  Medidas de dispersión

```python
# Varianza muestral

np.var(X)
```

```python
# Desviación típica muestral

std = np.std(X)
std
```

```python
# Coeficiente de variación std/mean*100
# Variabilidad relativa entre la media y la std, si hay mucha variabilidad el coeficiente será grande.

std/mean*100
```

```python
# 3rd Moment: Asimetría de Fisher

sk = sp.skew(X)
sk
```
