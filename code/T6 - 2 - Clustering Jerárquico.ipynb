# Clustering jerárquico y dendrogramas
Antes de empezar, pongamos un poco de notación para hablar todos el mismo idioma

* X dataset (array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos 
* Z array de enlace del cluster con la información de las uniones
* k número de clusters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import cophenet
from scipy.cluster.hierarchy import inconsistent
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]],size=[100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]],size=[50,])
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
Z = linkage(X,"ward")
Z
```

```python
c, coph_dist = cophenet(Z,pdist(X))
c
```

### Representación gráfica de un dendrograma

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma del clustering jerárquico")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z,color_threshold=0.7*180)
plt.show()
```

### Truncar el dendrograma

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma del clustering jerárquico truncado")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z,color_threshold=0.7*180, truncate_mode="lastp", p=12, show_leaf_counts=True, show_contracted=True)
plt.show()
```

### Dendrograma tuneado

```python
def dendrogram_tune(*args,**kwargs):        
    max_d = kwargs.pop("max_d", None)
    if (max_d and 'color_threshold' not in kwargs):
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above',0)
    
    ddata = dendrogram(*args,**kwargs)    
    
    if (not kwargs.get('no_plot', False)):
        plt.title("Clustering jerárquico con Dendrograma truncado")
        plt.xlabel("Índice del dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for index, distance, color in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(index[1:3])
            y = distance[1]
            if (y > annotate_above):
                plt.plot(x,y,'o',c=color)
                plt.annotate("%.3g"%y,(x,y),xytext=(0,-5),
                            textcoords = "offset points", va="top", ha="center")
    
    if (max_d):
        plt.axhline(y=max_d, c='k')
    
    return ddata
```

```python
dendrogram_tune(Z,truncate_mode="lastp",p=12,show_contracted=True,annotate_above=10,leaf_rotation=90.0,max_d=50)
plt.show()
```

inconsistency_i = (h_i-avg(h_j))/std(h_j)

```python
depth = 3
incons = inconsistent(Z,depth)
incons[-10:]
```

### Método del codo

```python
last = Z[-10:,2]
last_rev = last[::-1]
print(last_rev)
idx = np.arange(1,len(last)+1)
plt.plot(idx,last_rev)

acc = np.diff(last,2)
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1,acc_rev)
plt.show()
k = acc_rev.argmax()+2
print(f"el número óptimo de clusters es {k}")
```

```python
c = np.random.multivariate_normal([40,40],[[20,1],[1,30]], size=[200,])
d = np.random.multivariate_normal([80,80],[[30,1],[1,30]], size=[200,])
e = np.random.multivariate_normal([0,100],[[100,1],[1,100]], size=[200,])
X2 = np.concatenate((X,c,d,e),)
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
Z2 = linkage(X2,"ward")
```

```python
plt.figure(figsize=(10,10))
dendrogram_tune(
    Z2,
    truncate_mode="lastp",
    p=30,
    show_contracted=True,
    annotate_above = 40,
    max_d = 300
)
plt.show()
```

```python
last = Z2[-10:,2]
last_rev = last[::-1]
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)

acc = np.diff(last,2)
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1, acc_rev)
plt.show()
k = acc_rev.argmax() +2
print(f"El número óptimo de clusters es {k}")
```
